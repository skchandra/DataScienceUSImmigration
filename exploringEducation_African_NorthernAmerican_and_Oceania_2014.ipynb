```python
def make_piBar(things2Plot, labels, title):
    labels = labels
    sizes = things2Plot
    colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
    patches, texts = plt.pie(sizes, colors=colors, shadow=True, startangle=90)
    plt.legend(patches, labels, loc="best")
    plt.axis('equal')
    plt.tight_layout()
    plt.title(title)
    plt.show()
```

```python
def make_piBar_multiple(things2Plot1, things2Plot2, things2Plot3):
    plt.figure(figsize=(20, 7))
    ax1 = plt.subplot(1, 4, 1)
    labels = ['Nursery School', 'Elementary School', 'High School', 'College']
    colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
    patches= plt.pie(things2Plot1, colors=colors, autopct='%1.1f%%')
    plt.legend(patches, labels = labels, loc="best")
    plt.tight_layout()
    plt.title('People Born in Africa enrolled in School', fontsize = 15)
    
    ax2 = plt.subplot(1, 4, 2)
    colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
    patches= plt.pie(things2Plot2, colors=colors, autopct='%1.1f%%')
    plt.legend(patches, labels = labels, loc="best")
    plt.tight_layout()
    plt.title('People Born in North America enrolled in School', fontsize = 15)
    
    ax3 = plt.subplot(1, 4, 3)
    colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
    patches= plt.pie(things2Plot3, colors=colors, autopct='%1.1f%%')
    plt.legend(patches, labels = labels, loc="best")
    plt.tight_layout()
    plt.title('People Born in Oceania enrolled in School', fontsize = 15)
    
    
    plt.show()
```

```python
def many_bars(x, y1, y2, y3, y4, y5, title, xlabels): 
    x_decreased_by_01 = [y - 0.1 for y in x]
    x_decreased_by_02 = [y - 0.2 for y in x]
    x_increased_by_01 = [y + 0.1 for y in x]
    x_increased_by_02 = [y + 0.2 for y in x]
    print y1 
    print y2 
    print y3
    print y4 
    print y5
    ax = plt.subplot(111)
    bar1 = ax.bar(x_decreased_by_02, y1,width=0.1,color='b',align='center', label = 'less than hs')
    bar2 = ax.bar(x_decreased_by_01, y2,width=0.1,color='g',align='center', label = 'high school')
    bar3 = ax.bar(x, y3,width=0.1,color='r',align='center', label = 'some college')
    bar4 = ax.bar(x_increased_by_01, y4,width=0.1,color='c',align='center', label = 'bachelors degree')
    bar5 = ax.bar(x_increased_by_02, y5,width=0.1,color='y',align='center', label = 'graduate school')
    
    plt.legend(handles=[bar2, bar3, bar4])
    plt.title(title)
    plt.xticks(x, xlabels)
    plt.show()
```
