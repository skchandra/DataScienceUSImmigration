```python
import pandas 
import numpy
import matplotlib.pyplot as plt
import importingData
%matplotlib inline
```

```python
def make_piBar(things2Plot, labels, title):
    labels = labels
    sizes = things2Plot
    colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
    patches, texts = plt.pie(sizes, colors=colors, shadow=True, startangle=90)
    plt.legend(patches, labels, loc="best")
    plt.axis('equal')
    plt.tight_layout()
    plt.title(title)
    plt.show()
```

```python
def make_piBar_multiple(things2Plot1, things2Plot2, things2Plot3):
    plt.figure(figsize=(20, 7))
    ax1 = plt.subplot(1, 4, 1)
    labels = ['Nursery School', 'Elementary School', 'High School', 'College']
    colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
    patches= plt.pie(things2Plot1, colors=colors, autopct='%1.1f%%')
    plt.legend(patches, labels = labels, loc="best")
    plt.tight_layout()
    plt.title('Europeans enrolled in School', fontsize = 15)
    
    ax2 = plt.subplot(1, 4, 2)
    colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
    patches= plt.pie(things2Plot2, colors=colors, autopct='%1.1f%%')
    plt.legend(patches, labels = labels, loc="best")
    plt.tight_layout()
    plt.title('Northern and Western Europeans enrolled in School', fontsize = 15)
    
    ax3 = plt.subplot(1, 4, 3)
    colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
    patches= plt.pie(things2Plot3, colors=colors, autopct='%1.1f%%')
    plt.legend(patches, labels = labels, loc="best")
    plt.tight_layout()
    plt.title('Southern and Eastern Europeans enrolled in School', fontsize = 15)

    
    plt.show()
```

Now we want to compare the level of education attainment each type of European recieved. We want to know what type of degree they have and for people over 25, what their education background is like. We have this information and we can plot it below. 

First we have to collect the data of the education level of each region.

```python
ed_attainment_europe = europeanData.educ_attainment_born_in_Europe()
ed_attainment_northern_and_western_europe = europeanData.educ_attainment_born_in_Northern_and_Western_Europe()
ed_attainment_southern_and_eastern_europe = europeanData.educ_attainment_born_in_Southern_and_Eastern_Europe()
```

```python
X = numpy.arange(5)
width = 0.2
plt.figure(figsize=(15,11))
p1 = plt.bar([i-0.2 for i in X], ed_attainment_europe[1:6], color = '#073B4C', width=width)
p2 = plt.bar(X, ed_attainment_northern_and_western_europe[1:6], color='#9ACD32', width = width)
p3 = plt.bar([i+0.2 for i in X],ed_attainment_southern_and_eastern_europe[1:6], color='#F08080', width = width)
p4 = plt.xticks(X+width/2.,('< H.S edu','H.S education','Some college','Bachelors degree','Graduate Degree'), fontsize=16)
plt.xlabel('Degree', fontsize=16)
plt.ylabel('Percentage of European Foreigners earning these degrees of education', fontsize=16)
plt.title('Degrees Obtained By European Foreign Born Population', fontsize = 25)
plt.legend((p1[0], p2[0], p3[0], p4[0]), ('Born in Europe', 'Born in Northern or Western Europe', 'Born in Southern or Eastern Europe'), fontsize=12, loc = "best")
plt.show()
```

The graph above shows that the majority of Europeans from North or Western Europe have some college, and the second highest is a high school education. This graph also follows the Gaussian distribution. The peak for southern and eatern europeans is a high school education. Only around 18% of Europeans have a bachelors degree and slightly less of them have a graduate degree.
