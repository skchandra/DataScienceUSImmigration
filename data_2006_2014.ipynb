```python
def clean(x):
    if '%' in x:
        return float(x.strip('%'))   
    elif ',' in x:
        return float(x.replace(',',''))
    else:
        return float(x)
```

```python
def line_plot(data,title,xlabel,ylabel,xticks,legend1,legend2):
    plt.figure(figsize=(15,11))
    xvals = 0
    for i,val in enumerate(data):
        xvals = np.arange(len(val))
        plt.plot(np.arange(len(val)),val)
    plt.title(title,fontsize=16)
    plt.xlabel(xlabel,fontsize=16)
    plt.ylabel(ylabel,fontsize=16)
    plt.xticks(xvals,xticks,fontsize=12)
    plt.legend(legend1,legend2,fontsize=12)
    
    import matplotlib.ticker as tkr
    formatter = tkr.ScalarFormatter(useMathText=False)
    formatter.set_scientific(False)
    
    plt.show()
```
