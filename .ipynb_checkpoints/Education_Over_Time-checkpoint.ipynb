I got tired of copying and posting this code everywhere, so I made a line plot function. It takes in as input the data, title, labels, and legend.

```python
def line_plot(data,title,xlabel,ylabel,xticks,legend):
#     define figure size
    plt.figure(figsize=(15,11))
    xvals = 0
    p = {}
#     plotting
    for i,val in enumerate(data):
        xvals = numpy.arange(len(val))
        p[i] = plt.plot(numpy.arange(len(val)),val)
    plt.title(title,fontsize=16)
    plt.xlabel(xlabel,fontsize=16)
    plt.ylabel(ylabel,fontsize=16)
    plt.xticks(xvals,xticks,fontsize=12)
    legend1 = [p[i][0] for i in p]
    plt.legend(legend1,legend,fontsize=12)
    
    import matplotlib.ticker as tkr
    formatter = tkr.ScalarFormatter(useMathText=False)
    formatter.set_scientific(False)
    
    plt.show()
```

Below is a cleaning function. The goal of the cleaning function was to remove NaNs, make the content of every cell a float, and remove the percent signs. 

```python
def clean_dataFrame(data): 
    for i in range(len(data)): 
        row = data.iloc[i]
        for j in range(len(row)): 
            new_string = str(row.iloc[j]).strip('%')
            index = new_string.find(".")
            if (index > 0): ## float 
                row.iloc[j] = float(new_string) 
            elif(new_string.isdigit()): #integer
                row.iloc[j] = int(new_string)
    return data
```

# People age 16-24 enrolled in College or Graduate School 

Now we want to explore the rates of enrolling in college or graduate school. Throughout the years of 2006 to 2013, is there an increasing trend in foreign born immigrants enrolling in college or graduate school?

This is her clean function that strips away the percentage and turns everything into a float.

```python
def clean(x):
    try:
        if '%' in x:
            return float(x.strip('%'))   
        elif ',' in x:
            return float(x.replace(',',''))
        else:
            return float(x)
    except:
        return float(x)
```
