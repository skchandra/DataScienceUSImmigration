This notebook is for exploring trends over time in employment, income, and other pertinent factors from 2006-2014. The data was taken from the American Community Survey and is available in our repository.


```python
%matplotlib inline
#importing packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#open data
df_2006 = pd.read_table('2006.tsv')
df_2007 = pd.read_table('2007.tsv')
df_2008 = pd.read_table('2008.tsv')
df_2009 = pd.read_table('2009.csv')
df_2010 = pd.read_table('2010.csv')
df_2011 = pd.read_table('2011.csv')
df_2012 = pd.read_table('2012.csv')
df_2013 = pd.read_table('2013.tsv')
df_2014 = pd.read_table('2014.csv')
df_2009
```

```python
#basic cleaning function to make sure everything is the correct type
def clean(x):
    try:
        if '%' in x:
            return float(x.strip('%'))   
        elif ',' in x:
            return float(x.replace(',',''))
        else:
            return float(x)
    except:
        return float(x)
```

```python
#line plot function 
def line_plot(data,title,xlabel,ylabel,xticks,legend):
    plt.figure(figsize=(15,11))
    xvals = 0
    p = {}
    for i,val in enumerate(data):
        xvals = np.arange(len(val))
        p[i] = plt.plot(np.arange(len(val)),val)
    plt.title(title,fontsize=16)
    plt.xlabel(xlabel,fontsize=16)
    plt.ylabel(ylabel,fontsize=16)
    plt.xticks(xvals,xticks,fontsize=12)
    legend1 = [p[i][0] for i in p]
    plt.legend(legend1,legend,fontsize=12)
    
    import matplotlib.ticker as tkr
    formatter = tkr.ScalarFormatter(useMathText=False)
    formatter.set_scientific(False)
    
    plt.show()
```

```python
#plot unemployment rate over time

xticks = ('2006','2007','2008','2009','2010','2011','2012','2013','2014')
legend = ('Asia','Europe','Latin America','Africa','Northern America','Oceania')
title= 'Unemployment Rate from 2006-2014'
ylabel = 'Percentage of Group\'s Workforce'
xlabel = 'Years'

line_plot([unemployment_asia,unemployment_europe,unemployment_latin_america,unemployment_africa,unemployment_northern_america,unemployment_oceania],title,xlabel,ylabel,xticks,legend)
```
